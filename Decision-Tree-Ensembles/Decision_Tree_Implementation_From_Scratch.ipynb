```python
""" 
DECISION TREES - Implement Decision Tree Algorithms ID3, C4.5 and CART from scratch and 
apply it to classify handwritten digit images. This example is using the MNIST database 
of handwritten digits as training samples 

SCRATCH - I DID NOT rely on any packages to facilitate the implementation. The results 
presented below are generated by Decision Tree Algorithm ID3. 
"""
from __future__ import print_function

import numpy as np  
from decision_tree import Decision_Tree

from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/data/", one_hot=False)
train_xdata = np.array([np.reshape(x, (28, 28)) for x in mnist.train.images])
test_xdata = np.array([np.reshape(x, (28, 28)) for x in mnist.test.images]) 
train_data = np.reshape(train_xdata, (train_xdata.shape[0],train_xdata.shape[1]*train_xdata.shape[2])) 
test_data = np.reshape(test_xdata, (test_xdata.shape[0],test_xdata.shape[1]*test_xdata.shape[2]))
train_labels = mnist.train.labels
test_labels = mnist.test.labels 

print("train_data.shape = ", train_data.shape)

# Parameters
num_categories = 10
num_features = 784 # Each image is 28x28 pixels
max_depth_of_tree = 10
size = 5
batch_size = 10000
batch_x, batch_y = mnist.train.next_batch(batch_size) 

# Please note that key for root is set as 0
myDecisionTree = Decision_Tree(batch_x, batch_y, size, num_features, max_depth_of_tree, 0, num_categories) 

right = 0
wrong = 0
for i in range(test_data.shape[0]):
    y = myDecisionTree.decision_tree_prediction(test_data[i])
    if round(y) == test_labels[i]: 
        right += 1
    else: 
        wrong += 1

print("num of wrong preditions = ", wrong)
print("num of right predictions = ", right)
print("total num of validation samples = ", test_data.shape[0])
print("accuracy = ", float(right)/float(test_data.shape[0]))
```
